```python
#1
import time
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append(r'C:\OS\CA3\Part5')


import Cython1
import Cython2
import Cython3
import Cython4



def time_function(func, *args, **kwargs):
    start_time = time.time()
    func(*args, **kwargs)
    return time.time() - start_time

lattice = np.random.randint(2, size=(10000, 10000))


python_time = time_function(Python.update, lattice)
cython1_time = time_function(Cython1.update, lattice)
cython2_time = time_function(Cython2.update, lattice)
cython3_time = time_function(Cython3.update, lattice)
cython4_time = time_function(Cython4.update, lattice)


times = [python_time, cython1_time, cython2_time, cython3_time, cython4_time]
labels = ['Python', 'Cython1', 'Cython2', 'Cython3', 'Cython4']

#Graph
plt.bar(labels, times, color='blue')
plt.xlabel('Implementation')
plt.ylabel('Time (sec)')
plt.title('Results')
plt.show()
```

```python
#2
import numpy as np
import matplotlib.pyplot as plt

import Python
import Cython1
import Cython2
import Cython3
import Cython4



def time_function(func, *args, **kwargs):
    start_time = time.time()
    func(*args, **kwargs)
    return time.time() - start_time

lattice = np.random.randint(2, size=(100000, 100000))
i=1
j=1

python_time = time_function(Python.update_rule, lattice,i,j)
cython1_time = time_function(Cython1.update_rule, lattice,i,j)
cython2_time = time_function(Cython2.update_rule, lattice,i,j)
cython3_time = time_function(Cython3.update_rule, lattice,i,j)
cython4_time = time_function(Cython4.update_rule, lattice,i,j)

times = [python_time, cython1_time, cython2_time, cython3_time, cython4_time]
labels = ['Python', 'Cython1', 'Cython2', 'Cython3', 'Cython4']

#Graph
plt.bar(labels, times, color='blue')
plt.xlabel('Implementation')
plt.ylabel('Time (sec)')
plt.title('Results')
plt.show()
```
